```python
r = ['Breaking International News & Views', 'MarketBeat: Stock Market News and Research Tools',
'Examiner is back',
'Newsmax Finance | Business News - Financial Saving | Investing',                  
'HugeDomains.com',                                                                          
'USA TODAY' ,                                                                                  
'Tele Management – Latest TV shows news & reviews' ,                                           
'The Epoch Times'   ,                                                                          
'NewsFactor' ,                                                                                 
'interactive investor – the UK’s number one flat-fee investment platform'   ,                  
'Slate’s Use of Your Data' ,                                                                  
'GadgetellSolutions Media' ,                                                                   
'Article expired'    ,                                                                         
'BidnessETC'   ,                                                                                
'EUROPE ONLINE'  ,                                                                             
'ITPro Today: IT News, How-Tos, Trends, Case Studies, Career Tips, More',                      
'Take control of your money',                                                                   
'GamePolitics News'  ,                                                                          
'Breaking News & Headlines'  ,                                                                  
'Broadway Tickets & Videos' ,                                                                  
'Virtual Press Office: Tradeshow press release distribution, press kits, event marketing',      
'Home Page'  ,                                                                                 
'Opinion and Commentary']

repeated_rows = []
small_title = []
for index, row in df.iterrows():
    if df['Title'][index] in r:
        repeated_rows.append(index)
    tokens = nltk.word_tokenize(str(df['Title'][index]))
    if len(tokens) < 5:
        repeated_rows.append(index)
    art_tokens = nltk.word_tokenize(str(df['Article'][index]))
    if len(art_tokens) <200:
        small_title.append(index)

```

```python
import numpy as np
doc_lengths = []

for article in articles:
    # get rough token count distribution
    tokens = nltk.word_tokenize(article)
    doc_lengths.append(len(tokens))
doc_lengths = np.array(doc_lengths)
```
